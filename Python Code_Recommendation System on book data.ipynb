### EDA

```python
book.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
book_df= book.rename({'User.ID':'userid','Book.Title':'booktitle','Book.Rating':'bookrating'},axis=1)
book_df.head()
```

```python
book_df[book_df.duplicated()]
```

```python
book_df= book_df.drop_duplicates(subset=['userid','booktitle'])
book_df
```

#### Number of Unique users in the dataset

```python
len(book_df.userid.unique())
```

```python
len(book_df.booktitle.unique())
```

```python
user_book_df = book_df.pivot(index='userid',columns='booktitle',values='bookrating').reset_index(drop=True)
```

```python
user_book_df
```

```python
user_book_df.index = book_df.userid.unique()
```

```python
user_book_df
```

```python
user_book_df.fillna(0,inplace=True)
```

```python
user_book_df
```

```python
# import libraries to find similarity
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

### Similarities between two readers

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

###### Set the index and column names to user ids 

```python
user_sim_df.index = book_df.userid.unique()
user_sim_df.columns = book_df.userid.unique()
```

```python
user_sim_df.iloc[0:10,0:10]
```

```python
np.fill_diagonal(user_sim,0)
user_sim_df.iloc[0:10,0:10]
```
