```python
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
movie = pd.read_csv("C:\\Users\\jhishna2022\\Downloads\\my_movies.csv")
movie
```

#### EDA

```python
movie1=movie.drop(['V1','V2','V3','V4','V5'],axis=1)
movie1
```

### Apriori Algorithm

```python
frequent_itemsets = apriori(movie1, min_support = 0.1, use_colnames=True)
frequent_itemsets
```

### Association Rule for minimum confidence of 0.6

```python
rules = association_rules(frequent_itemsets, metric= 'confidence', min_threshold = 0.6)
rules
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns
sns.scatterplot(rules['support'],rules['confidence'])
```

```python
sns.scatterplot(rules['confidence'],rules['lift'])
```

```python
sns.scatterplot(rules['support'],rules['lift'])
```

```python
rules[rules.lift>1]
```
