```python
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
book=pd.read_csv("C:\\Users\\jhishna2022\\Downloads\\book.csv")
book
```

```python
book.info()
```

## Apriori Algorithm

```python
frequent_itemsets = apriori(book, min_support = 0.1, use_colnames = True)
frequent_itemsets
```

### Association rule for minimum confidence of 0.6

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
rules
```

```python
rules[rules.lift>1]
```

```python
rules['lift'].max()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns
sns.scatterplot(rules['support'],rules['confidence'])
```

```python
sns.scatterplot(rules['support'],rules['lift'])
```

```python
sns.scatterplot(rules['confidence'],rules['lift'])
```

```python
rules.sort_values('lift',ascending = False)
```
